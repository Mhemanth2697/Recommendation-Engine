```python
def getHTMLPage(url):
    response = requests.get(url, allow_redirects = True, timeout = 5)
    #meta = dict(response.headers)
    content = response.text
    return content
def parseContent(content):
    soup = BeautifulSoup(content, 'lxml')
    g_data = soup.find_all("a", {"class": "actionLinkLite bookPageGenreLink"})
    return g_data
def getGenres(isbn):
    temp = []
    if not(isinstance(isbn, str)):
        return []
    url1 = "https://www.goodreads.com/book/isbn/"+isbn+"?format=FORMAT"
    try:
        genreList = parseContent(getHTMLPage(url1))
    except:
        return []
    for i in genreList:
        i = str(i)
        ind = i.find("href=\"/genres/")
        t = i[ind:]
        g = t[(t.find("\">"))+2:]
        temp.append(g[:len(g)-4])
    if len(genreList) == 0:
        return["too", "few", "ratings"]
    return temp
```
